```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import csv
import time

# --- Configuration Chrome/Edge ---
chrome_options = Options()
chrome_options.add_argument("--headless")  # Mode headless
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

# --- Configuration Selenium ---
options = Options()
# options.add_argument("--headless")  # Décommente pour ne pas afficher la fenêtre
options.add_argument("--window-size=1920x1080")

driver_path = r"C:\Users\PC\.wdm\drivers\chromedriver\chromedriver-win64\chromedriver.exe"
driver = webdriver.Chrome(service=Service(driver_path), options=options)
URL = "https://www.transfermarkt.fr/wettbewerbe/europa/wettbewerbe?plus=1"

driver.get(URL)
time.sleep(10)

championnats = []
id_counter = 1

while True:
    # Récupérer les lignes du tableau
    # Sélection de toutes les cellules qui contiennent le lien du championnat
    cells = driver.find_elements(By.CSS_SELECTOR, "td.hauptlink")

    for cell in cells:
        links = cell.find_elements(By.TAG_NAME, "a")
        if len(links) >= 2:
            link = links[1]  # le 2e <a> contient le nom du championnat
            nom = link.text.strip()
            href = link.get_attribute("href")
            championnats.append((id_counter, nom, href))
            id_counter += 1
            print(nom, href)

    # Vérifier si le bouton "page suivante" existe et est cliquable
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR, "a.tm-pagination__link[title='Allez à la page suivante']")
        next_btn.click()
        time.sleep(2)  # Attendre le chargement de la nouvelle page
    except:
        break  # Pas de page suivante, on sort

driver.quit()

# Sauvegarder dans CSV
with open("championnats_europe.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["ID", "Championnat", "Lien"])
    for row in championnats:
        writer.writerow(row)

print(f"{len(championnats)} championnats récupérés et sauvegardés dans championnats_europe.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import csv
import time

def parse_value_euro(valeur_str):
    """Convertit la valeur en euros entiers"""
    if not valeur_str:
        return 0
    valeur_str = valeur_str.replace("€", "").strip()
    valeur_str = valeur_str.replace(".", "").replace(",", ".")
    try:
        number = float(valeur_str.split()[0])
        if "mrd" in valeur_str:
            return int(number * 1_000_000_000)
        elif "mio" in valeur_str:
            return int(number * 1_000_000)
        else:
            return int(number)
    except:
        return 0

# --- Configuration Selenium ---
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--window-size=1920x1080")

driver_path = r"C:\Users\PC\.wdm\drivers\chromedriver\chromedriver-win64\chromedriver.exe"
driver = webdriver.Chrome(service=Service(driver_path), options=chrome_options)

URL = "https://www.transfermarkt.fr/wettbewerbe/europa/wettbewerbe?plus=1"
driver.get(URL)
time.sleep(10)  # attendre le chargement complet

championnats = []
id_counter = 1
seen = set()  # pour éviter les doublons
page_count = 0
max_pages = 10

while page_count < max_pages:
    # Récupérer tous les championnats sur la page
    rows = driver.find_elements(By.CSS_SELECTOR, "table.items tbody tr")
    for row in rows:
        # Nom et lien du championnat
        links = row.find_elements(By.CSS_SELECTOR, "td.hauptlink a")
        if len(links) >= 2:
            link = links[1]
            nom = link.text.strip()
            href = link.get_attribute("href")
        else:
            continue  # ligne invalide

        # Éviter doublons
        if (nom, href) in seen:
            continue
        seen.add((nom, href))

        # Valeur du championnat
        try:
            valeur_td = row.find_element(By.CSS_SELECTOR, "td.rechts.hauptlink")
            valeur_i = valeur_td.text.strip()
        except:
            valeur = ""
        valeur = parse_value_euro(valeur_i)

        championnats.append((id_counter, nom, href, valeur))
        print(id_counter, nom, href, valeur)
        id_counter += 1

    page_count += 1

    # Pagination : chercher le bouton "page suivante"
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR, "a.tm-pagination__link[title='Allez à la page suivante']")
        next_btn.click()
        time.sleep(10)  # attendre le chargement complet de la page
    except:
        break  # pas de page suivante, on sort de la boucle

driver.quit()

# Sauvegarder dans CSV
with open("championnats_europe.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["ID", "Championnat", "Lien", "Valeur"])
    for row in championnats:
        writer.writerow(row)

print(f"{len(championnats)} championnats récupérés et sauvegardés dans championnats_europe.csv")
```
